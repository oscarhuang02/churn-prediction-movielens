```python
# Load the data from the data folder
data_path = 'data/movies.csv'  # Replace 'your_file_name.csv' with the actual file name
data = pd.read_csv(data_path)

# Display the first few rows of the data
print(data.head())

data.describe()
```

```python
# convert the timestamp to a datetime object
data['timestamp'] = pd.to_datetime(data['timestamp'], unit='s')
```

```python
# average rating and count of ratings per user
user_ratings = data.groupby(['userId']).agg({'rating': ['count', 'mean']}).reset_index()
```

```python
# average rating per movie
movie_ratings = data.groupby(['movieId']).agg({'rating': ['count', 'mean']}).reset_index()
movie_ratings
```
